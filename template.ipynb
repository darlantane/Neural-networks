```python
#import
import numpy as np
import pandas as pd
```

```python
#dataset
df = pd.read_csv("train.csv")
df
```

```python
# détection
df.isnull().sum()
```

```python
# supprime les lignes avec NaN
df.dropna()
# supprime les colonnes
df.dropna(axis=1)
```

```python
# Suppression des doublons
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# imputer
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")
df[["age", "salary"]] = imputer.fit_transform(df[["age", "salary"]])
```

```python
# Types de données
df['Outstanding_Debt'] = pd.to_numeric(df['Outstanding_Debt'], errors='coerce')
df['Annual_Income'] = pd.to_numeric(df['Annual_Income'], errors='coerce')
```

```python
#Sélection
df = df[['Outstanding_Debt', 'Annual_Income', 'Credit_Score']]
```

```python
#num
df = df[df['Credit_Score'] != 'Standard']
df['Credit_Score'] = df['Credit_Score'].map({
    'Good': 0,
    'Poor': 1
})
```

```python
# Valeurs aberrantes
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)
IQR = Q3 - Q1

df = df[(df["salary"] >= Q1 - 1.5 * IQR) &
        (df["salary"] <= Q3 + 1.5 * IQR)]

# Chaines de caractères
df["city"] = df["city"].str.strip()
df["city"] = df["city"].str.lower()
df["city"] = df["city"].str.replace(r"[^a-z]", "", regex=True)
```

```python
#features and target
X = df[['Outstanding_Debt', 'Annual_Income']].values
y = df['Credit_Score'].values
```

```python
#normalisation
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
#encodage et standardisation
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

num_features = ["age", "salary"]
cat_features = ["city", "gender"]

num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("num", num_pipeline, num_features),
    ("cat", cat_pipeline, cat_features)
])
```

```python
#Classification binaire
import tensorflow as tf
from tensorflow.keras import Sequential, layers
from tensorflow.keras.layers import Input

def initialize_model():
    model = Sequential()
    model.add(Input(shape=(2,)))
    model.add(layers.Dense(20, activation='relu'))
    model.add(layers.Dense(10, activation='relu'))
    model.add(layers.Dense(5, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    
    model.compile(
        loss='binary_crossentropy', 
        optimizer='adam', 
        metrics=['accuracy'])
    return model

model = initialize_model()

model.summary()
```

```python
#CNN – Classification d’images (MNIST)
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train[..., None] / 255.0
X_test = X_test[..., None] / 255.0

def initialize_model():
    model = Sequential([
        Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
        MaxPooling2D((2,2)),
        Conv2D(64, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Flatten(),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy'])
    return model

model = initialize_model()

model.summary()
```

```python
history = model.fit(
    X_train, y_train, 
    epochs=5)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy test : {accuracy:.2f}")
```

```python
# NLP – Classification de texte
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
    "j'adore ce produit",
    "c'est horrible",
    "excellent service",
    "très mauvais"
]
labels = [1, 0, 1, 0]

tokenizer = Tokenizer(num_words=1000)
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
X = pad_sequences(sequences, maxlen=5)

def initialize_model():
    model = Sequential([
        Embedding(1000, 16),
        LSTM(16),
        Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

model = initialize_model()
model.summary()
```

```python
history = model.fit(
    X, 
    labels, 
    epochs=10)
```
